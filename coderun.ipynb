```python
data = data.dropna(subset=['description'])
```

```python
data.fillna("No description available", inplace=True)  # for the entire DataFrame
```

```python
import re

# Define a simple list of common stopwords
stopwords = set(["and", "the", "is", "in", "to", "of", "a", "for", "on", "with", "as", "by", "at", "it", "from", "this", "that"])

# Preprocessing function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Remove stopwords
    text = ' '.join(word for word in text.split() if word not in stopwords)
    return text

# Apply preprocessing to 'product_name' and 'description'
data['product_name'] = data['product_name'].apply(preprocess_text)
data['description'] = data['description'].apply(preprocess_text)

# Display to confirm
print(data[['product_name', 'description']].head())
```

```python
#feature engineering

from sklearn.feature_extraction.text import TfidfVectorizer
# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer(max_features=500)  # Adjust max_features as needed

# Fit and transform 'description' column
tfidf_matrix = vectorizer.fit_transform(data['description'])

# Convert to a DataFrame for easier viewing (optional)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
print(tfidf_df.head())
```
